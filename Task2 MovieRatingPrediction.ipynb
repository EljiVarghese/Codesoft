```python
top_movie = movie_data.loc[movie_data['Rating'].sort_values(ascending=False).head(10).index]
top_movie
```

```python
sns.histplot(data=top_movie,x='Year',hue='Rating',multiple='stack')
plt.title("Top 10 Movies")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.show()
```

```python
genre_counts = movie_data['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre','Count']

top_n_generes = genre_counts.head(5)
top_n_generes
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Genre',y='Count',data=top_n_generes)
plt.title('Top 5 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show
```

```python
director_avg_rating = movie_data.groupby('Director')['Rating'].mean().reset_index()
director_avg_rating.columns = ['Director','AVG Rating']
director_avg_rating = director_avg_rating.sort_values(by='AVG Rating',ascending=False).head(10)
director_avg_rating
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='AVG Rating',y='Director',data=director_avg_rating)
plt.title('Top 10 Directors')
plt.xlabel('AVG Rating')
plt.ylabel('Directors')
plt.show()
```
